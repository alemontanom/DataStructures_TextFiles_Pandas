#### Problema 1  (20 puntos)


Desarrolla una funcion que reciba los datos de habitantes y la superficie del territorio de un estado para calcular su densidad de poblacion, la funcion debe entonces determinar si la densidad es baja, media, alta o extrema

Si el indice es:

    <50 la densidad es baja
    de 50 y hasta 140 la densidad es media
    entre 141 y 500 la densidad es alta
    superior a 500 la densidad es extrema

```python
def densidad(h, t):
    d=h/t
    if d<50:
        return("baja")
    elif d>=50 and d<=140:
        return("media")
    elif d>140 and d<=500:
        return("alta")
    else :
        return("extrema")
```
