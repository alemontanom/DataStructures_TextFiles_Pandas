# Ejercicios

Para los equipos que perdieron mas de 50 partidos compara en barras el total de goles a favor y en contra

```python
mun[mun.PartidosPerdidos>50].pivot_table(index="Equipo", values=["GolesAFavor", "GolesEnContra"])
```

```python
mun[mun.PartidosPerdidos>50].pivot_table(index="Equipo", values=["GolesAFavor", "GolesEnContra"]).plot(kind="bar")
```

Grafica el porcentaje de equipos que participan de cada pais

```python
mun.groupby("Pais").Pais.count().plot(kind="pie", subplots=True, autopct="%1.2f%%", explode=[.1]*10)
```

Para Estados Unidos, Mexico y Espana compara los Goles a Favor y En contra

```python
mun.query("Pais=='USA' or Pais=='Mexico' or Pais=='Espana'").pivot_table(index="Pais", values=["GolesAFavor", "GolesEnContra"], aggfunc="sum").plot(kind="bar")
```

```python
mun.query('(Pais=="USA" or Pais=="Mexico" or Pais=="Espana")').groupby("Pais")[["GolesAFavor","GolesEnContra"]].sum().plot(kind="bar")
```

```python
n=mun.pivot_table(index="Pais",values=["GolesAFavor","GolesEnContra"],aggfunc="sum")
n.loc[["Mexico", "USA", "Espana"]].plot(kind="bar")
```
