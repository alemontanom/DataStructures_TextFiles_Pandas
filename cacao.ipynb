### ¿cuantos registros de critica existen por pais? considera el pais de origen de la empresa 

### recuerda que cada renglon del dataframe es una critica

```python
import pandas as pd
```

Aleeee tienes que abrir el archivo cacao.csv en excel y guardarlo con formato .xlsx para que funcione esto. Si abres el csv te da un problema del formato de caracteres y asi.

```python
cacao=pd.read_excel("cacao.xlsx")
cacao
```

```python
cacao.groupby("CompanyLocation")["Rating"].count()
```

Segunda Opción

```python
cacao.pivot_table(index="CompanyLocation",values="Rating",aggfunc="count")
```

### ¿cual es la calificacion (Rating) promedio por año? 

```python
cacao.pivot_table(index="ReviewDate\n",values="Rating")
```

```python
cacao.groupby("ReviewDate\n")["Rating"].mean()
```

### ¿cuanto suman la calificaciones  por cada pais de origen del grano? 

```python
cacao.pivot_table(index="Broad BeanOrigin",values="Rating",aggfunc="sum")
```

### ¿cual es el rating promedio de las barras francesas ? 

```python
cacao.query('CompanyLocation=="France"')["Rating"].mean()
```

### ¿cuantas criticas se hicieron para la barra llamada Belize? 

### recuerda que cada renglon del dataframe es una critica

```python
cacao.query('Name=="Belize"')["Rating"].count()
```

### entre las barras que tienen entre 60 y 80 porciento de cacao muestra su calificacion (rating) promedio, maximo y minimo por pais de origen del grano de cacao 

```python
cacao.query('CocoaPercent>=0.60 and CocoaPercent<=0.80').pivot_table(index="BroadBeanOrigin",values="Rating",aggfunc=["mean","max","min"])
```

### Crea una tabla que muestre entre las empresas de origen Estadounidenses que utilizan los granos tipo Criollo o Forastero muestra la calificacion promedio  y maxima que han dado a las barras  por tipo de grano y porcentaje de cacao 

### Utiliza pivot table

```python
usacriolloforastero=cacao.query('CompanyLocation=="U.S.A." and (BeanType=="Criollo" or BeanType=="Forastero")')
```

```python
usacriolloforastero.pivot_table(index=["BeanType","CocoaPercent"], values="Rating",aggfunc=["sum","max"],margins=True)
```

```python
usacriolloforastero.pivot_table(index="CocoaPercent",columns="BeanType",values="Rating",aggfunc=["sum","max"],margins=True)
```
